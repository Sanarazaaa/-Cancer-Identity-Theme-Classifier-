```python
import pandas as pd

# Simulated mini dataset of patient statements
data = {
    "text": [
        "I’ve accepted cancer as part of my journey and grown stronger.",
        "I learned so much from my experience and want to help others.",
        "I can’t stop thinking about the pain. I feel like I’m still stuck in it.",
        "I don’t want to be seen as a cancer patient anymore. It’s not who I am.",
        "I talk openly about my treatment. It’s part of who I am now.",
        "The whole experience has made me appreciate life more deeply.",
        "Even after treatment, I still feel like a patient.",
        "I just want to forget everything related to cancer."
    ],
    "label": [
        "Acceptance",
        "Enrichment",
        "Engulfment",
        "Rejection",
        "Acceptance",
        "Enrichment",
        "Engulfment",
        "Rejection"
    ]
}

df = pd.DataFrame(data)
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Split data
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.25, random_state=42)

# Create pipeline
model = make_pipeline(TfidfVectorizer(), MultinomialNB())

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Report
print(classification_report(y_test, y_pred))
ConfusionMatrixDisplay.from_estimator(model, X_test, y_test)
plt.show()
```

```python
# Let user try it
print("Type a cancer-related statement to classify it:")
text = input()

prediction = model.predict([text])[0]
print(f"\nPredicted Cancer Identity Theme: **{prediction}**")
```
